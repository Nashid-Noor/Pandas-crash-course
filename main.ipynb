## Example of how to use data frame

```python
import pandas as pd
data={
    'Name': ['Nash','Noor','Nin','Fas'],
    'Age': [21,22,10,50],
    'Country': ['IND','UK','Africa','China']
}
df= pd.DataFrame(data) 
print(df)
```

## Column wise Indexing 

```python
df["Name"]
```

```python
df[["Name","Age"]]
```

## Row-wise indexing

```python
df.iloc[0]
```

```python
print(df.iloc[0][0])
print(df.iloc[0][1])
print(df.iloc[0]["Country"])
```

## Filtering

```python
print(df[df["Name"]=="Nash"])
print("########################")
print(df[df["Age"]>20])
print("########################")
print(df[df["Country"].isin(["IND"])])
print("########################")
print(df[(df["Country"]=="IND") & (df["Age"]>20)])
```

## Updating

```python
df['Age']=df['Age']+5
print(df)
df['Country'=="IND"]="India"
print(df)
df.loc[df["Name"]=="Nash","Name"]="Nashid"
print(df)
```

## Deleting

```python
df=df.drop(0)
df
```

## Clean up data

```python
df.dropna()
df.fillna({"Name":"No Name"},inplace=True)
df
```

## Data Analysis

```python
print(df["Country"].value_counts())
print("###########################")
print(df["Age"].mean()) 
print("###########################")
print(df.groupby("Country")["Age"].sum())
```

## To save modified data to a csv file

```python
df.to_csv("modified.csv",index=False)
df.to_csv("modified_withindex.csv",index=True)
```
